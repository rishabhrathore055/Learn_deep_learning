## TF 2 and Keras : Sequential API

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.utils import to_categorical, plot_model
from tensorflow.keras.datasets import mnist
```

```python
print(tf.__version__)
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
plt.imshow(x_train[1],cmap='gray',interpolation=None)
print(y_train[1])
```

```python
plt.imshow(x_train[0],cmap='gray',interpolation=None)
print(y_train[0])
```

```python
# number of labels
num_labels = len(np.unique(y_train))
print(num_labels)
```

```python
# convert y_1's to one-hot vectors
y_train = to_categorical(y_train)
y_test= to_categorical(y_test)
```

```python
# Image 2d matrix to 10 vectors
img_size = x_train.shape[1]
inputSize = img_size * img_size

# 2d to 1d Normalization (0-255)
x_train = np.reshape(x_train,[-1,inputSize])
x_train = x_train.astype('float32')/255
x_test = np.reshape(x_test,[-1,inputSize])
x_test = x_test.astype('float32')/255
```

```python
# Hyper parameters of Neural network
batch_size = 128
hidden_units = 256
```

```python
# Sequential API

# 3 layer Multilayer preceptron ReLu
model = Sequential()
model.add(Dense(hidden_units,input_dim = input_size))
model.add(Activation('relu'))
model.add(Dense(hidden_units))
model.add(Activation('relu'))
model.add(Dense(num_labels))

# this is the output for one hot vector
model.add(Activation('softmax'))
model.summary()
```

```python
model.compile(loss ='categorical_crossentropy',
             optimizer = 'adam',
             metrics =['accuracy'])
```

```python
# train the network
%time
tf.config.run_functions_eagerly(True)
model.fit(x_train,y_train,epochs=20,batch_size = batch_size)
```

```python
# testing the dataset performance
loss, acc = model.evaluate(x_test,y_test,
                          batch_size= batch_size,
                          verbose=0)
print("Test accuracy %.1f"%(100.0 *acc))
print(loss)

# Overfit or underfit
```

## TF 2 and Keras : Functional API

```python
print(x_train.shape)
```

```python
inputs = keras.Input(shape=(784,))
inputs.shape
```

```python
# Functional API : much more intuitive
x1 = layers.Dense(hidden_units,activation='relu')(inputs)
x2 = layers.Dense(hidden_units,activation='relu')(x1)
outputs = layers.Dense(num_labels,activation='softmax')(x2)

model = keras.Model(inputs=inputs,outputs = outputs, name = "simple model")
model.summary()
```

```python
model.compile(optimizer='adam',
             loss = 'categorical_crossentropy',
             metrics=['accuracy'])
```

```python
tf.config.run_functions_eagerly(True)
model.fit(x_train,y_train,epochs=20,batch_size = batch_size)
```

```python
# testing the dataset performance
loss, acc = model.evaluate(x_test,y_test,
                          batch_size= batch_size,
                          verbose=0)
print("Test accuracy %.1f"%(100.0 *acc))
print(loss)

# Overfit or underfit
```
