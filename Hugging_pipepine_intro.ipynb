###  Zero Shot Classification

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "this is a course about the transformers library",
    cadidate_labels = ["education","politics","business"]
)
```

### Text Genration

```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation")
generator("In this tutorial, I want to learn about")
```

### With gpt2

```python
from transformers import pipeline
generator = pipeline("text-generation",model= "distilgpt2")
generator(
        "In this tutorial, I want to learn about",
          max_length = 30,
          num_return_sequences = 2)
```

```python
from transformers import pipeline
ner = pipeline("ner",grouped_entities=True)
ner("My name is rishabh rathore and I am  data science and machine learning learner")
```

### Question Answering

```python
from transformers import pipeline
question_answer = pipeline("question-answering")
question_answer(
    question = "Where do I study?",
    context = "I'm rishabh and I study in indore institute of science and technology"
)
```
